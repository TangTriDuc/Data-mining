# 3. Basic Matplotlib Commands


```python
plt.plot(x,y,'g')
plt.xlabel('x here')
plt.ylabel('y here')
plt.title('string')
plt.show()
```

# 4. Creating Multiplots on Same Canvas



```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-');
```

# 5. Object Oriented Method


```python
# Create Figure (empty canavas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# Plo
axes.plot(x,y,'g')
axes.set_xlabel('Set X')
axes.set_ylabel('Set Y')
axes.set_title('Set Title')
```

```python
# Creates blanks canavas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# Larger Figure Axes 1
axes1.plot(x,y,'r')
axes1.set_xlabel('Set X_axes2')
axes1.set_ylabel('Set Y_axes2')
axes1.set_title('Set Title Axes2')

# Insert Figure Axes 2
axes2.plot(x,y,'g')
axes2.set_xlabel('Set X_axes2')
axes2.set_ylabel('Set Y_axes2')
axes2.set_title('Set Title Axes2')
```

# 6. Suplot()


```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and ax
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('Set X')
axes.set_ylabel('Set Y')
axes.set_title('Set Title')
```

```python
# Empty canavas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Axes is an array of axes to plot on
axes
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
    fig
    plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
    fig
    plt.tight_layout()
```

# 11. Legends


```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

# 12. Plot range


```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0, 60])
axes[2],set_xlim([2, 5])
axes[2],set_title("custom axes range");
```

# II. Seaborn

# 1. Load testing dataset

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
%matplotlib inline

sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

# 2. Scatter plot


```python
ax = sns.scatterplot(x="total_bill", y = "tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", hue="sex", size="size")
```

# 3. Categorical functions

```python
sns.catplot(x="sex", y="total_bill", hue="day", data=tips, kind="strip")
```

```python
sns.catplot(x="sex", y="total_bill", hue="day", data=tips, kind="strip", jitter=False, dodge= True)
```

```python
sns.catplot(x="sex", y="total_bill", hue="day", data=tips, kind="box")
```
